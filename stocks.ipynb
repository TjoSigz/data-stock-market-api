**Import libraries**



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**Make call to URL**

```python
import requests

url = "https://iex.lewagon.com/stable/stock/aapl/chart/3m"
api_data = requests.get(url).json()
```

**Create apple_df DataFrame**

```python
apple_df=pd.DataFrame(api_data)
apple_df.head(4)
```

**Convert the date to a datetime object**

```python
apple_df['priceDate'] = pd.to_datetime(apple_df['priceDate'],format="%Y-%m-%d")
apple_df.info()
```

**Set the date column as the index**

```python
#apple_df = apple_df.set_index('date')
apple_df.head(5)
```

***Plot with only the values in the close column**

```python
close_df = apple_df['close']
close_df.plot()
```

**Plot with the values in the open, close, high, and low columns**

```python
apple_df[['open', 'close', 'high', 'low']].plot(figsize= (12,5))
```
